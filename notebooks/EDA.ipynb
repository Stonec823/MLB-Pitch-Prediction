Now that we have an initial featureset after preprocessing, I'll explore relationships and feature significane in the data.

The goal by the end of the notebook is to have a final featureset before the modeling and validation phase.

```python
from pybaseball import  playerid_lookup
from pybaseball import  statcast_pitcher
from pybaseball import  statcast
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/raw/chris_sale.csv')
df.head()
sns.countplot(data=df, x='inning', hue='pitch_type')
plt.title('Class Distribution by Feature')
plt.show()
```

```python
df = df.drop(columns=(['game_pk', 'pitch_type']))
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12);
# save heatmap as .png file
# dpi - sets the resolution of the saved image in dots/inches
# bbox_inches - when set to 'tight' - does not allow the labels to be cropped
plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
```
